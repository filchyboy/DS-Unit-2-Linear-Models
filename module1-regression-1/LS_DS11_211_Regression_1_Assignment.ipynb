###Cats

Created a simple function to run the value of having access to cats  when seeking an apartment in NYC. Of interest to me would be how this might change with multiple cats and/or where in the city the rental is to be found.

```python
cat = 0
cat_conundrum = LinearRegression()
features = ['cats_allowed']
target = 'price'
X_train = df[features]
y_train = df[target]
cat_conundrum.fit(X_train, y_train)
X_test = [[cat]]
y_pred = cat_conundrum.predict(X_test)
y_pred
cat_conundrum.coef_
cat_conundrum.intercept_
mae = mean_absolute_error(y_test, y_pred)
def predict(cat):
    y_pred = cat_conundrum.predict([[cat]])
    estimate = y_pred[0]
    coefficient = cat_conundrum.coef_[0]

    if cat > 0:
      explainer = f"{cat:,.0f} cat adds ${coefficient:,.0f}. Our model's error: ${mae:,.0f}"
    else:
      explainer = f"{cat:,.0f} cat means no additional cost. Our model's error: ${mae:,.0f}"
      
    result = f'${estimate:,.0f} estimated monthly price to rent an apartment in NYC.'
    explanation = f'{cat:,.0f} cat adds ${coefficient:,.0f}.'
    return result + '\n' + explainer

```

```python
print(predict(1))
```
